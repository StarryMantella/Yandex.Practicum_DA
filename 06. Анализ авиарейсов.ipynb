**–í —Ç–∞–±–ª–∏—Ü–µ Query_3 –æ–∫—Ä—É–≥–ª–∏–º —á–∏—Å–ª–∞ –≤ —Å—Ç–æ–ª–±—Ü–µ average_flights –¥–æ 2-—Ö –∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π, –∏ –∏–∑–º–µ–Ω–∏–º –Ω–∞ —Ç–∏–ø 'float':**
