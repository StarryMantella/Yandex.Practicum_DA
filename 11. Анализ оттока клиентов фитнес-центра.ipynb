–ú–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –ø–∞—Ä–∞–º–µ—Ç—Ä stratify = y –≤ —Ñ—É–Ω–∫—Ü–∏–∏ train_test_split( ), –≥–¥–µ y ‚Äì —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∞—è —Ü–µ–ª–µ–≤–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è. –î–∞–Ω–Ω—ã–µ —Ä–∞–∑–¥–µ–ª—è—Ç—Å—è –Ω–∞ –¥–≤–∞ –∫–ª–∞—Å—Ç–µ—Ä–∞ (–Ω–∞–±–ª—é–¥–µ–Ω–∏—è –∫–ª–∞—Å—Å–∞ 0 –∏ –Ω–∞–±–ª—é–¥–µ–Ω–∏—è –∫–ª–∞—Å—Å–∞ 1), –∏ –≤ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é –≤—ã–±–æ—Ä–∫—É –ø–æ–ø–∞–¥—É—Ç –ø–æ 20% –Ω–∞–±–ª—é–¥–µ–Ω–∏–π –∏–∑ –∫–∞–∂–¥–æ–≥–æ –∫–ª–∞—Å—Ç–µ—Ä–∞:


  
–¢–æ –µ—Å—Ç—å, –µ—Å–ª–∏ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤ –≤ –æ—Ä–∏–≥–∏–Ω–∞–ª—å–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ –±—ã–ª–æ 2:5, —Ç–æ –≤ –æ–±—É—á–∞—é—â–µ–π –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∞—Ö –æ–Ω–æ —Å–æ—Ö—Ä–∞–Ω–∏—Ç—Å—è —Ç–∞–∫–∏–º –∂–µ. –¢–∞–∫–æ–π –ø–æ–¥—Ö–æ–¥ –≤ —á–µ–º-—Ç–æ —Ä–µ—à–∞–µ—Ç –ø—Ä–æ–±–ª–µ–º—É –Ω–µ—Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –∫–ª–∞—Å—Å–æ–≤.

–§—É–Ω–∫—Ü–∏—è train_test_split( ) —Å—Ç—Ä–æ–µ–Ω–∞ —Ç–∞–∫, —á—Ç–æ –º—ã –Ω–µ —Å—Ç–∞–ª–∫–∏–≤–∞–µ–º—Å—è —Å –ø–æ–¥–æ–±–Ω–æ–π –ø—Ä–æ–±–ª–µ–º–æ–π. –ù–æ —è –≤—Å–µ –∂–µ –ø–∏—à—É –ø—Ä–æ —ç—Ç–æ—Ç –ø–∞—Ä–∞–º–µ—Ç—Ä, —á—Ç–æ–±—ã –æ–±—Ä–∞—Ç–∏—Ç—å –≤–Ω–∏–º–∞–Ω–∏–µ –Ω–∞ –ø—Ä–æ–±–ª–µ–º—É –¥–∏—Å–±–∞–ª–∞–Ω—Å–∞ :)</div>

```python
X = data.drop(['churn'], axis = 1)
y = data['churn']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
scaler = StandardScaler()
X_train_st = scaler.fit_transform(X_train)
X_test_st = scaler.transform(X_test)
```

**3.2  –û–±—É—á–∏—Ç–µ –º–æ–¥–µ–ª—å –Ω–∞ train-–≤—ã–±–æ—Ä–∫–µ –¥–≤—É–º—è —Å–ø–æ—Å–æ–±–∞–º–∏:<br>
–õ–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–µ–π:**

```python
lr_model = LogisticRegression(random_state=0)
lr_model.fit(X_train_st, y_train)
lr_probabilities = lr_model.predict_proba(X_test_st)[:,1]
lr_predictions = lr_model.predict(X_test_st)

print('Accuracy: {:.3f}'.format(accuracy_score(y_test, lr_predictions)))
print('Precision: {:.3f}'.format(precision_score(y_test, lr_predictions)))
print('Recall: {:.3f}'.format(recall_score(y_test, lr_predictions)))
print('F1: {:.3f}'.format(f1_score(y_test, lr_predictions)))
```

**–°–ª—É—á–∞–π–Ω—ã–º –ª–µ—Å–æ–º:**

```python
rf_model = RandomForestClassifier(n_estimators = 100, random_state = 0) 
rf_model.fit(X_train_st, y_train)
rf_probabilities = rf_model.predict_proba(X_test_st)[:,1]
rf_predictions = rf_model.predict(X_test_st)

print('Accuracy: {:.3f}'.format(accuracy_score(y_test, rf_predictions)))
print('Precision: {:.3f}'.format(precision_score(y_test, rf_predictions)))
print('Recall: {:.3f}'.format(recall_score(y_test, rf_predictions)))
print('F1: {:.3f}'.format(f1_score(y_test, rf_predictions)))
```
